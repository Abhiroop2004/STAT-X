## Importing Libraries

```python
import pandas as pd
import numpy as np
```

## Importing Data

```python
df = pd.read_csv("Kol-1973.csv", index_col=[1], parse_dates=[1])
df.head(10)
```

### Checking Stations

```python
df['STATION'].value_counts()
```

### Checking Duplicate Index

```python
df.index.value_counts()
```

## Filter Dataset for station 42809099999

```python
df2 = df.loc[(df['STATION'] == 42809099999)]
df2.head(10)
```

```python
df2.info()
```

## Filter Data further to only include data from 1973 onwards

```python
data = df2.loc['1973-01-01':]
data.head()
```

## Preprocess this data

```python
for i in range(len(data.columns)):
    print(f"Value counts for column {data.columns[i]}:")
    print(data.iloc[:, i].value_counts())  # value_counts on each column
    print()
```

```python
data.drop(columns=['STATION','GUST', 'SNDP', 'STP', 'SLP'], inplace=True)
data.head(10)
```

### Setting Invalid Observations as NaN
Originally set as 99.9 or 999.9

```python
pd.set_option('display.max_rows', None) 
```

```python
data.loc[data['PRCP'] == 99.99, 'PRCP'] = np.nan
data.loc[data['MAX'] == 9999.9, 'MAX'] = np.nan
data.loc[data['MIN'] ==  9999.9, 'MIN'] = np.nan
data.loc[data['DEWP'] == 9999.9, 'DEWP'] = np.nan
data.loc[data['MXSPD'] == 999.9, 'MXSPD'] = np.nan
data.loc[data['VISIB'] == 999.9, 'VISIB'] = np.nan
data.loc[data['WDSP'] == 999.9, 'WDSP'] = np.nan
```

```python
data['FRSHTT'].value_counts()
```

### After Analyzing only 3 weather parameters have sufficient occurrence in the dataset (>1%)
They are Fog, Rain and Thunder

```python
conditions = ['Fog', 'Rain', 'Snow', 'Hail', 'Thunder', 'Tornado']
data[conditions] = data['FRSHTT'].apply(lambda x: pd.Series(list(f"{int(str(x), 2):06b}"), dtype=int))  # Convert to 6-bit binary string
data[conditions].value_counts()
```

```python
data.head()
```

```python
print(data['Rain'].value_counts())
print(data['Snow'].value_counts())
print(data['Fog'].value_counts())
print(data['Hail'].value_counts())
print(data['Thunder'].value_counts())
print(data['Tornado'].value_counts())
```

```python
data.drop(columns=['FRSHTT'], inplace=True)
data.head(10)
```

## Fixing Missing values with Interpolation

### If Precipitation > 0.0 is marked as No-Rain day, mark it as Rainy day

```python
data.loc[(data['Rain'] == 0) & (data['PRCP'] > 0.0), 'Rain'] = 1.0
```

### If No-Rain day has precipitation as NaN, set it as 0.0

```python
data.loc[(data['Rain'] == 0) & (data['PRCP'].isna()), 'PRCP'] = 0.0
```

```python
data = data.asfreq(fill_value=None, freq='1d')
data.head(10)
```

```python
data.isnull().sum()
```

```python
data['Rain'].value_counts()
```

```python
df_interpolated = data.interpolate(method='time')
df_interpolated.head(10)
```

```python
df_interpolated[conditions] = data[conditions].fillna(0)
```

```python
df_interpolated.head()
```

### If 0 Precipitation is marked as Rainy day, mark it as No-Rain day and if >0 Precipitation is marked as No-Rain day, mark it as Rainy day

```python
df_interpolated.loc[(df_interpolated['Rain'] == 1.0) & (df_interpolated['PRCP'] == 0.000000), 'Rain'] = 0.0
```

```python
df_interpolated.loc[(df_interpolated['PRCP'] == 0.000000) & (df_interpolated['Rain'] == 1.0)]
```

## Converting to Metric Units

```python
df_interpolated[['MAX', 'MIN', 'TEMP']] = df_interpolated[['MAX', 'MIN', 'TEMP']].apply(lambda x: (x - 32) * 5 / 9)
df_interpolated['PRCP'] = df_interpolated['PRCP'] * 25.4
```

## Removing 0 Values from Min

```python
df_interpolated[df_interpolated['MIN'] == 0].index
```

```python
df_interpolated.loc[['1977-01-03', '2000-12-04'], 'MIN'] = np.nan
```

```python
df_interpolated = df_interpolated.interpolate(method='time')
```

```python
df_interpolated.to_csv('Kolkata-1973-2024-cleaned.csv', index = True)
```
